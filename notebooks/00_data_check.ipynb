```python
# Sanity check: confirm the M5 data loads and looks right
import pandas as pd
from pathlib import Path

# Project root is one level up from notebooks/
DATA_DIR = Path("..") / "data" / "raw"

print("Files in data/raw/:")
for f in DATA_DIR.iterdir():
    size_mb = f.stat().st_size / (1024 * 1024)
    print(f"  {f.name}: {size_mb:.1f} MB")
```

```python
# Load each file and peek at it
sales = pd.read_csv(DATA_DIR / "sales_train_evaluation.csv")
calendar = pd.read_csv(DATA_DIR / "calendar.csv")
prices = pd.read_csv(DATA_DIR / "sell_prices.csv")

print(f"sales:    {sales.shape[0]:,} rows × {sales.shape[1]:,} cols")
print(f"calendar: {calendar.shape[0]:,} rows × {calendar.shape[1]:,} cols")
print(f"prices:   {prices.shape[0]:,} rows × {prices.shape[1]:,} cols")
```

```python
# Peek at the first few rows of each
print("SALES (first 3 rows, first 10 cols)")
print(sales.iloc[:3, :10])
print("\n CALENDAR (first 3 rows)")
print(calendar.head(3))
print("\n PRICES (first 3 rows)")
print(prices.head(3))
```
